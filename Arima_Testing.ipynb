```python
"""
THIS MODULE CONTAINS THE CODES REQUIRED THE GENERATE THE CHARTS FOR THE NHMIS BULLETIN
"""
import csv
import pandas as pd
import requests
import slack

# Google Sheet interactions
import gspread
from google.oauth2 import service_account
from google.auth.transport.requests import AuthorizedSession
from gspread_dataframe import set_with_dataframe, get_as_dataframe


# Apache Airflow Server details or Google Drive Location



class DataGenerator:
	"""
	This class is responsible for the generation of the data for the NHMIS Bulletin
	"""

	def __init__(self):
		self.links = "https://dhis2nigeria.org.ng/dhis/api/29/analytics.csv?dimension=dx:HVZTC0vlhW9;O82o1WlMisO.REPORTING_RATE_ON_TIME;O82o1WlMisO.REPORTING_RATE;Jc1WjNKrObY;KWJ3cSuyzs4;SCDCoP6l9vz;k2Ncm24FnEA;dCQZ2WnhEwE;HH0iQLlqddM;UVw96W1kcja;YjZiHDKMWCJ;CvX3NUIe50P;rgmSg4j5CRD;HhWjY78dECn;iWQ9dIazJD4;bKSVofVt1vy;BFeSKtHhPKi;m5gTF1jKQjl;GEd2F6skCpT;G7iWnz9RMy9;mz0LIaWBAz9.k5pHlvkYR3T;mz0LIaWBAz9.jaAWWXYY0mP;mz0LIaWBAz9.HBvhntn3sam;wGPpop3rz7i;rXoqTMxWvfR&dimension=pe:LAST_6_MONTHS&dimension=ou:s5DPBsdoE8b;LEVEL-1&displayProperty=NAME&tableLayout=true&columns=dx&rows=pe;ou"
		self.map_links = "https://dhis2nigeria.org.ng/dhis/api/29/analytics.csv?dimension=dx:mz0LIaWBAz9.HBvhntn3sam;mz0LIaWBAz9.k5pHlvkYR3T;mz0LIaWBAz9.jaAWWXYY0mP&dimension=pe:LAST_MONTH&dimension=ou:s5DPBsdoE8b;LEVEL-1;LEVEL-2&displayProperty=NAME&tableLayout=true&columns=dx&rows=pe;ou"
		self.bulletin_link = "https://docs.google.com/spreadsheets/d/1IFXvFLIkJx0PXhhc0pjJ5KdSQ1dVygTEcAtF3e4lRgY/edit#gid=273934337"
		# self.credentials_path = f"{BASE_FOLDER}/extract_data.json"
		# self.scope = ["https://www.googleapis.com/auth/spreadsheets"]
		# self.credentials = service_account.Credentials.from_service_account_file(
		# 	self.credentials_path, scopes=self.scope)
		# self.scoped_credentials = self.credentials.with_scopes(
		# 	self.scope
		# )
		self.worksheet_id = "1IFXvFLIkJx0PXhhc0pjJ5KdSQ1dVygTEcAtF3e4lRgY"
		self.extracted_chart_data = pd.DataFrame()
		self.all_data = pd.DataFrame()

	@staticmethod
	def extract_data(links):
		"""
		  This function is used to extract the report straight from DHIS2
		  Args:
			  links: NHMIS urls containing links to the data

		  Returns:
			  The required data
		  """
		csv_file_path = f"msdat_data.csv"

		def create_csv_file(new_data):
			"""
			Function to write downloaded data into csv for easy  manipulation

			output:
			- A csv file containing data downloaded

			"""
			# create an empty csv file

			data_file = open(csv_file_path, 'w', newline='')

			# create the csv writer object
			csv_writer = csv.writer(data_file, delimiter='\t')

			for lines in new_data:
				# writing data to csv file
				csv_writer.writerow([lines])

			# close file object
			data_file.close()

		response = requests.get(links,
		                        auth=("FMOH_ePHD", "@e4eDHIS2"))
		data = response.content

		# decode from bytes to string and then lists
		new_data = data.decode('utf-8').splitlines().copy()

		create_csv_file(new_data)
		# read new csv file created
		new_df = pd.read_csv(csv_file_path)

		return new_df

	def extract_chart_data(self):
		self.extracted_chart_data = self.extract_data(self.links)
		unwanted = ['periodid', 'periodcode', 'perioddescription', 'organisationunitid', 'organisationunitname',
		            'organisationunitcode',
		            'organisationunitdescription']

		self.extracted_chart_data.drop(unwanted, axis=1, inplace=True)
		self.extracted_chart_data.to_csv(f'extracted_data.csv', index=False)

	# def export_data(self):
	# 	"""
	# 	This is used to export data to Google Sheet

	# 	data_retrieved: DataFrame object
	# 	sheet_name: Name of the sheet that the data is being uploaded to
	# 	"""
	# 	clean_data = pd.read_csv(f'{BASE_FOLDER}/clean_data.csv',
	# 	                         index_col=None)
		
	# 	gc = gspread.Client(auth=self.scoped_credentials)
	# 	gc.session = AuthorizedSession(self.scoped_credentials)
	# 	worksheet_object = gc.open_by_key(self.worksheet_id)

	# 	# Upload data to sheet
	# 	data_sheet = worksheet_object.worksheet("bulletin_data")
	# 	set_with_dataframe(data_sheet, clean_data)

	def transform_data(self):
		"""
		Transforms the extracted data to the format needed for the bulletin.
		The output is also exported to a Google Sheet for reference
		Returns : None
		"""
		extracted_data = pd.read_csv(f'extracted_data.csv',
		                             index_col=None)

		all_dataframes = []
		for col in extracted_data.columns[1:]:
			new_dataframe = pd.DataFrame(columns=['Period',
			                                      'Organisation',
			                                      'Indicator',
			                                      'Value',
			                                      'Source',
			                                      'Indicator_datasource'])

			new_dataframe['Period'] = extracted_data['periodname']
			new_dataframe['Organisation'] = "National"
			new_dataframe['Indicator'] = col
			new_dataframe['Value'] = extracted_data[col]
			new_dataframe['Source'] = 'NHMIS'
			new_dataframe['Indicator_datasource'] = f"{col}->NHMIS"
			all_dataframes.append(new_dataframe)

		self.all_data = pd.concat(all_dataframes)
		


	def extract_opv(self):
		"""
		Groups the Different OPVs i.e OPV0, OPV1 by months and aggregate them by their sum.
		Append the resulting dataframe back the filtered dataframe

		"""
	
		 #Filters out entries that have OpV andentries that dont have OPV
		df_no_opv = self.all_data[~self.all_data['Indicator_datasource'].str.contains("OPV")] 
		df_new_2 = self.all_data[self.all_data['Indicator_datasource'].str.contains("OPV")]

		groupby_df= df_new_2.groupby('Period', as_index = False)['Value'].sum() 
		groupby_df["Organisation"]= "National"   #Adding the other column to match the main df
		groupby_df["Indicator"]= "OPV Birth"
		groupby_df["Source"]= "NHMIS"
		groupby_df["Indicator_datasource"]= "OPV->NHMIS"

		final_df= pd.concat([df_no_opv, groupby_df], axis=0)
		final_df.reset_index(inplace=True, drop=True,)

		final_df.to_csv(f'clean_data.csv', index=False)

	# def send_slack_message(self):
	# 	"""
	# 	This function is used to send slack messages to the channel when the data has been updated.
	# 	"""
	# 	token = "xoxb-9535526721-3000281800049-pWCaWJ4HAywWrADrOLVtjadO"
	# 	slack_message = f"""
	# 	    		        Hello,

	# 	    		        The NHMIS Bulletin Sheet has been updated.

	# 	    		        Please verify data update by using the link below:

	# 	    		        <{self.bulletin_link}|*NHMIS Bulletin Data Update*>

	# 	    		        Regards,

	# 	    		        PyData Team
	# 					"""
	# 	client = slack.WebClient(token=token)

	# 	client.chat_postMessage(channel="#dacissh-discuss",
	# 	                        text=slack_message)


```
